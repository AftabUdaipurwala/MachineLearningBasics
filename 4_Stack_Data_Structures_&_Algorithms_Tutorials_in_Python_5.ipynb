```python
class Stack:
    def __init__(self):
        self.container = deque()
    
    def push(self,val):
        self.container.append(val)
    
    def pop(self):
        return self.container.pop()
    
    def peek(self):
        return self.container[-1] # see last element

    def is_empty(self):
        return len(self.container)==0
    
    def size(self):
        return len(self.container)
    
    def reverse(self):
        rlst = self.container.reverse()
        return rlst
```

```python
s = Stack()
s.push(5)
s.push(15)
s.push(52)
s.push(53)
```
