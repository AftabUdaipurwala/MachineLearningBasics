```python
monthly_expenses = [2200,2350,2600,2130,2190]
```

image.png

```python
# Q1 : Extra dollars spent in Feb compared to jan
extra = monthly_expenses[1]-monthly_expenses[0]
print(extra)
```

```python
#Q2 : Total Expenses in 1st 3 months:
total=0
for i in range(0,3):
    total+=monthly_expenses[i]
print(total)
```

```python
#Q3: When 2000 Dollars was spent?
spent=2000
for i in range(0,len(monthly_expenses)):
    if spent==monthly_expenses[i]:
        print(f"Monthly expenses in {i}th month was $2000")
```

```python
# Q4: Add june months expenses $1980 
monthly_expenses.append(1980)
print(monthly_expenses)
```

```python
#Q5: refund amount $200
monthly_expenses[3]-=200
print(monthly_expenses)
```
