# Create Decorator

```python
import time
def time_it(function):
    def wrapper(*args,**kwargs):
        start = time.time()
        result = function(*args,**kwargs)
        end=time.time()
        print(f'{function.__name__} took {(end-start)*1000}milisecs')
        return result
    return wrapper
```

# Bubble sort function

```python
@time_it
def bubble_sort(elements):
    size = len(elements)
    for j in range(size-1): # run this for each element in the list
        swapped = False
        for i in range(size-1-j): # compare that element with each other element in the list
            if elements[i]>elements[i+1]:
                elements[i],elements[i+1]=elements[i+1],elements[i]
                swapped=True
        if not swapped:
            print("list is already sorted, hence breaking")
            break
    return elements

```

# Evaluation

```python
if __name__=="__main__":
    lst=[1,45,85,65,74,25,31,54,63,41,15,23]
    sorted_list =bubble_sort(lst)
    print(sorted_list)
```

```python
lst = [1,2,3,4,5]
print(bubble_sort(lst))
```

```python
lst = ['a','c','w','p']
print(bubble_sort(lst))
```

```python
@time_it
def bubble_sort(elements,key):
    size = len(elements)
    for j in range(size-1): # run this for each element in the list
        swapped = False
        for i in range(size-1-j): # compare that element with each other element in the list
            if elements[i][key]>elements[i+1][key]:
                elements[i][key],elements[i+1][key]=elements[i+1][key],elements[i][key]
                swapped=True
        if not swapped:
            print("list is already sorted, hence breaking")
            break
    return elements

```
