```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv('/content/drive/MyDrive/py-master/ML/FeatureEngineering/3_outlier_IQR/heights.csv')
df.head(4)
```

```python
df.describe()
```

```python
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
Q1,Q3
```

```python
IQR= Q3-Q1
```

```python
LSL= Q1-1.5*IQR
USL= Q3+1.5*IQR
```

```python
df2 = df[(df.height>LSL)&(df.height<USL)]
```
